### lets create a plotting boundary

```python
import numpy as np
```

```python
def plot_decision_boundary(model,X,y):
    """
    Plots the desicion boundary created by a model prediction on x
    """
    
    # define the x boundaries of the plot and create a meshgrid
    x_min , x_max = X[:, 0].min() - 0.1 , X[:,1].max() + 0.1
    y_min , y_max = X[:, 1].min() - 0.1 , X[:,1].max() + 0.1
    xx, yy = np.meshgrid(np.linspace(x_min,x_max,100),np.linspace(y_min,y_max,100))
    
    # create x values (prediction by using these values)
    x_in = np.c_[xx.ravel(),yy.ravel()]
    
    # make predictions
    y_pred = model.predict(x_in)
#     print(f"before reshaping ypred is {y_pred.shape}")
#     print(y_pred)
    #check for the multicalss
    if model.output_shape[-1] > 1:
        print("This is a multiclass classification")
        
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
#         print(f"after reshaping ypred(mC) is {y_pred.shape}")
    else:
        print("This is a Binary Classsification")
        
        y_pred = np.round(np.max(y_pred, axis=1)).reshape(xx.shape)
#         print(f"after reshaping ypred(BC) is {y_pred.shape}")
#         print(y_pred)
        
    # plot the decision boundary
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model=model_3, X=X ,y=y)
```

# what is a sigmoid function?

```python
# let's create a tensor (linear)
A = tf.cast(tf.range(-10,10),tf.float32)
A
```

```python
# Let's plot the tensor
plt.plot(A)
```

```python
# let's see the sigmoid function and create our function
# The sigmoid function is - sigmoid(x) = 1 / (1 + exp(-x))
def sigmoid(x):
    return 1 / (1 + tf.exp(-x))
```

```python
# let's create sigmoid of tensor A
sig_a = sigmoid(A)
```

```python
# let's plot the sigmoid(A)
plt.plot(sig_a)
```

# What is Relu function?

```python
# relu is basically - max(0,x)
def relu(x):
    return tf.maximum(0, x)
```

```python
A
```

```python
rel_a = relu(A)
rel_a
```

```python
# plot the relu of tensor A
plt.plot(rel_a)
```

# what is Tanh function?

```python
# tanh function formula - ((exp(x) - exp(-x))/(exp(x) + exp(-x))).
def tanh(x):
    return ((tf.exp(x) - tf.exp(-x))/(tf.exp(x) + tf.exp(-x)))
```

```python
# create tanh of A
tan_a = tanh(A)
tan_a
```

```python
# let's plot the tanh function
plt.plot(tan_a)
```

## Finding the best learning rate

To find the ideal learning rate ( the learning rate where the loss decreases the most during training ) we're going to use the following steps:
* A learning rate **callback** - callback is an extra pair of functionality which we can add to the model and monitors while the model is training 👀.
* Another model (we could use the same model as above , but we are praticing building models here.
* A modified loss curves plot

```python
# set random seed
tf.random.set_seed(42)

# create a model 
model_8 = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

# compile a model
model_8.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

# create a callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch: 1e-4 * 10**(epoch/20))

# fit a model
history_8 = model_8.fit(x_train,y_train,epochs=100,callbacks=[lr_scheduler])
```

```python
pd.DataFrame(history_8.history).plot(figsize=(10,7),xlabel="epochs")
```

# plot the learning rate vs loss

```python
lrs = 1e-4 * (10**(tf.range(100)/20))
plt.figure(figsize=(10,7))
plt.semilogx(lrs, history_8.history["loss"])
plt.xlabel("Learning Rate")
plt.ylabel("Loss")
plt.title("Learning Rate Vs Loss", fontsize=16)
```

# Confusion Matrix

There are few terms in **confusion matrix** such as:
* **True Positives**: model predicts 1 when truth is 1
* **True Negatives**: model predicts 0 when truth is 0
* **False Positivites**: model predicts 1 when truth is 0
* **False Negativies**: model predicts 0 when truth is 1

<img src="images/confusion matrix.png"/>

```python
# import the confusion matrix from sklearn
from sklearn.metrics import confusion_matrix
```

```python
# predictions by using model_10
y_preds = model_10.predict(x_test)
```

```python
# get the confusion matrix now
# confusion_matrix(y_test, y_preds) # confusion matrix requires (y_true , y_preds)

# This gives us ValueError: Classification metrics can't handle a mix of binary and continuous targets
# so here y_test : have in binary format
# while y_preds is in prediction probability  form  which came from sigmoid activation function 
# so we have to convert y_preds to the binary format
# Let's do that :)
```
